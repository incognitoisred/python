# Matplotlib in Python

### Define Matplotlib : Matplotlib is a low level graph plotting library in python that serves as a visualization utility.

```python
import matplotlib
print(matplotlib.__version__)
```

### plot() function is used to draw points in a diagram

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints=np.array([0,10])
ypoints=np.array([0,400])

plt.plot(xpoints,ypoints)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints=np.array([2,10])
ypoints=np.array([4,12])

plt.plot(xpoints,ypoints)
plt.show()
```
